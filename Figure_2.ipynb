```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set font
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"

# initialize subplots
f, ax = plt.subplots(1,1, figsize=(5,3.5)) 
plt.subplots_adjust(wspace=0, hspace=0, top = 0.97, bottom = 0.13, right = 0.96, left = 0.14)
ax.tick_params(direction="in", top=True, right=True, length=5, labelsize=11.5)
ax.tick_params(which="minor", direction="in", top=True, right=True, length=3, labelsize=11.5)

# plot the downsampling fraction
logM = np.arange(10,15,0.1)
y = np.minimum(0, 2*(logM-12))
plt.plot(10**logM, 10**y)

# set axis limits and labels
plt.xlim(5e10,5e14)
plt.ylim(1e-3, 2)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$M_\mathrm{halo} \ / \ h^{-1}\mathrm{M}_\odot$", fontsize=11.5)
plt.ylabel("$f$", fontsize=11.5)

# save the figure as a pdf
plt.savefig("pdf/Figure_2.pdf", dpi=300)
```
